
>   concatenate data from different paths using merge and join functions to do the below : <br>
    1. outer join <br>
    2. inner join <br>
    3. left join<br>
    4.right join

```python
import pandas as pd
```

### Reading Files

```python
user_device = pd.read_csv('user_device.csv')
user_usage = pd.read_csv('user_usage.csv')
```

```python
print(user_device.shape)
user_device.head()
```

```python
print(user_usage.shape)
user_usage.head()
```

### 1. outer join 

```python
#with merge
pd.merge(user_device,user_usage,on='use_id',how='outer')
```

```python
#with join 
user_device.join(user_usage, on='use_id', how='outer', lsuffix='_device', rsuffix='_usage', sort=False).drop(['use_id'],axis = 1)
```

### 2. inner join

```python
#with merge
pd.merge(user_device,user_usage,on='use_id',how='inner')
```

```python
#with join 
user_device.set_index('use_id').join(user_usage.set_index('use_id'), on='use_id', how='inner').reset_index()
```

### 3. left join 

```python
#merge 
pd.merge(user_device,user_usage,on='use_id',how='left')
```

```python
#join
user_device.set_index('use_id').join(user_usage.set_index('use_id'), on='use_id', how='left').reset_index()
```

### 4. right join 

```python
#merge
pd.merge(user_device,user_usage,on='use_id',how='right')
```

```python
#join
user_device.set_index('use_id').join(user_usage.set_index('use_id'), on='use_id', how='right').reset_index()
```
