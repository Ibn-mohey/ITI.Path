```python
from dash.dependencies import Output,Input
import plotly.express as px
import pandas as pd
import dash
from dash import html
from dash import dcc
import dash
from jupyter_dash import JupyterDash
from dash import html
from dash import dcc 

df = px.data.gapminder()
fig = px.scatter(data_frame=df,x='gdpPercap',y='lifeExp',title='World',size_max=60,animation_frame='year',width=600,height=600,
           color='continent',log_x=True,size='pop',hover_name='country')

app = JupyterDash(external_stylesheets=["https://codepen.io/chriddyp/pen/bWLwgP.css"])

app.layout=html.Div(children=[
                        html.H1("Hello World!",style={'textAlign':'center','color':'red'}),
                        html.Div([
                            html.Div([dcc.Input(id='input_',
                                            placeholder='Enter a value...',
                                            type='text',
                                            value='')],className='six columns')
                        ]),html.Div(id='output_')
                        ],className='row')




@app.callback(Output(component_id='output_',component_property='children'),
             Input(component_id='input_',component_property='value'))
def boring_function(myinput):
    return myinput


app.run_server()
```
