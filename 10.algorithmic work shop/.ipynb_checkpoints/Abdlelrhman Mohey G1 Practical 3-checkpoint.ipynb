## Write a report Describing the Longest Increasing Subsequence (LIS), and define the problem and the solution 
Then implement the solution 

Given an integer array nums, return the length of the longest strictly increasing subsequence.

A subsequence is a sequence that can be derived from an array by deleting some or no elements without changing the order of the remaining elements. For example, [3,6,2,7] is a subsequence of the array [0,3,1,6,2,2,7].

```python
def LIS(array):

    length = [1]*len(array)
 
    for i in range(1, len(array)):
        for j in range(0, i):
            if array[i] > array[j] and length[i] <= (length[j] + 1):
                length[i] = length[j]+1
    print(length)             
    return max(length)

print ("test 1", LIS([10, 22, 9, 33, 21, 50, 41, 60])) #5
print ("test 2", LIS([0,8,4,12,2,10,6,14,1,9,5,13,3,11,7,15])) #6
print ("test 2", LIS([0,8,4,12,2,10,6,14,1,9,5,13,3,11,7,15,5])) #6
print ("test 3",LIS([10,9,2,5,3,7,101,18])) #4
print ("test 4",LIS( [7,7,7,7,7,7,7]) )#1
print ("test 5",LIS(list(range(0,5))[::-1]))#1
```
