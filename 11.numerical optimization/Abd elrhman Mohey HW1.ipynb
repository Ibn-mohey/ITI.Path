Use python code to build a function that calculate the approximate solution of a function using Newton’s method. <br>
a. Think about function arguments. **Function it self , x<sub>0</sub> to start and acceptable different from 0**<br> 
b. Choose the suitable stop criteria.  <br>

```python
def Newton(f,df,x_o,Acceptable_diff = 1e-5):
    count = 0
    x1 = x_o
    x2 = round(x1-(f(x1)/df(x1)),6)
    
    while abs(x1-x2)>=Acceptable_diff:
        x1 = x2
        x2 = round(x1-(f(x1)/df(x1)),6)
        print(x1)
        count +=1
    
    return x2,count
```

```python
def f1(x):
    return x**2-x-1

def df1(x):
    return 2*x-1
```

```python
answers = Newton(f1,df1,1)
print("Answer =", answers[0]," iterations =",answers[1] )
```

```python
def f2(x):
    return x**3-7*x**2+8*x-3

def df2(x):
    
    return 3*x**2-2*7*x+8
answers = Newton(f2,df2,5)
print("Answer =", answers[0]," iterations =",answers[1] )
```

```python
import math
def f3(x):
    return x*math.cos(x)-x**2

def df3(x):
    return math.cos(x)-x*math.sin(x)-2*x
answers = Newton(f3,df3,1)
print("Answer =", answers[0]," iterations =",answers[1] )
```
