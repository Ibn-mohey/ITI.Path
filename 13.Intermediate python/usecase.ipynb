```python
from sklearn.model_selection import train_test_split
def split_data(data):
    
    Y=np.array(data["Target"]).astype("float32")
    X=np.array(data.drop(["Target"],axis=1)).astype("float32")
    
    np.random.seed(0)
    
    train_features,test_features,train_labels,test_labels=train_test_split(X,Y,test_size=0.2)
    
    print("features =",X,"\n \n target= ", Y)
    
    return train_features,test_features,train_labels,test_labels
    
```

```python
 np.random.seed(0)
```

```python
train_features,test_features,train_labels,test_labels = split_data(new_data)
```

```python
print(f"Length of train_features is: {train_features.shape}")
print(f"Length of train_labels is: {train_labels.shape}")
print(f"Length of test_features is: {test_features.shape}")
print(f"Length of test_labels is: {test_labels.shape}")
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(train_features)
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)
```

## Model training and evaluation
  
  #### 2] define the model 
  #### 3] training 
  #### 4] prediction
  #### 5] Evaluation

```python
from sklearn import svm
from sklearn.model_selection import cross_val_score
```

```python
def get_scores(X_train,y_train,classifier):
    scores = cross_val_score(
    classifier,
    X_train,
    y_train,
    cv=10,
    scoring='accuracy',
    n_jobs=-1)
    return(scores)

get_scores(train_features,train_labels,svm.SVC()).mean()
```

```python
svm_model=svm.SVC(kernel='linear')
svm_model.fit(train_features,train_labels)

y_pred=svm_model.predict(test_features)
```

```python
from sklearn import metrics
from sklearn.metrics import f1_score,confusion_matrix, mean_squared_error
from sklearn.metrics import roc_auc_score, roc_curve, precision_score, recall_score

print(f" accuracy of training ={svm_model.score(train_features,train_labels)}")

print(f" accuracy of testing ={svm_model.score(test_features,test_labels)}")

print(f" Pression ={precision_score(y_pred,test_labels)}")
print(f" Recall ={recall_score(y_pred,test_labels)}")
print(f" f1 score ={f1_score(y_pred,test_labels)}")

svm_matrix = confusion_matrix(test_labels,y_pred)
plt.figure(figsize=(8, 8))
sns.heatmap(svm_matrix, annot=True,cbar=False, cmap="Blues", fmt="d")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix ')
```
