##  Q2:   get the local min of the function below using scipy. then plot the equation with the local min point using matlibplot


```python
from scipy.optimize import minimize 
```

```python
def FX(x):
    return x**2+x+2
opt = minimize(FX,0)
```

```python
opt.x
```

```python
x = np.linspace(-1,1,50)
plt.plot(x,FX(x),c='b');
plt.plot(opt.x,FX(opt.x),'dr');
```

```python
x = np.linspace(-10,10,50)
plt.plot(x,FX(x),c='b');
plt.plot(opt.x,FX(opt.x),'dr');
```

## Q4 use the best interpolation method to get Y_new  for this data 

        x = np.arange(0,360,5)
        y = np.sin(np.deg2rad(x))**2-np.cos(np.deg2rad(x))**2
        
   &    
        x_new = np.arange(0,360,10)



```python
x = np.arange(0,360,5)
y = np.sin(np.deg2rad(x))**2-np.cos(np.deg2rad(x))**2
x_new = np.arange(0,360,10)
```

```python
interp_function = interp1d(x,y)
y_newinterp1d = interp_function(x_new)
plt.plot(x,y,"o", x_new ,y_newinterp1d,".");
```

```python
plt.plot(x,y,"o", x_new ,y_newinterp1d);
```

```python
from scipy.interpolate import UnivariateSpline
inter_func2 = UnivariateSpline(x,y)
y_newUnivariateSpline1 = inter_func2(x_new)
plt.plot(x,y,"o", x_new ,y_newUnivariateSpline1);
```

```python
inter_func21 = UnivariateSpline(x,y)
inter_func21.set_smoothing_factor(.5)
y_newUnivariateSpline2 = inter_func21(x_new)
plt.plot(x,y,"o", x_new ,y_newUnivariateSpline2);
```

```python
inter_func21 = UnivariateSpline(x,y)
inter_func21.set_smoothing_factor(5)
y_newUnivariateSpline3 = inter_func21(x_new)
plt.plot(x,y,"o", x_new ,y_newUnivariateSpline3);
```

```python
from scipy.interpolate import Rbf
inter_func3 = Rbf(x,y)
y_newRbf = inter_func3(x_new)
plt.plot(x,y,"o", x_new ,y_newRbf);
```

```python
fig,ax = plt.subplots(2,2, figsize=(15,10))
fig
ax[0,0].plot(x,y,"o", x_new ,y_newinterp1d);
ax[0,0].set_title("Using interp1d");

ax[0,1].plot(x,y,"o", x_new ,y_newUnivariateSpline1);
ax[0,1].set_title("Using UnivariateSpline  with NO smoth");

ax[1,0].plot(x,y,"o", x_new ,y_newUnivariateSpline2);
ax[1,0].set_title("Using UnivariateSpline with smoth .5");

ax[1,1].plot(x,y,"o", x_new ,y_newRbf);
ax[1,1].set_title("Using Rbf");
```

```python
print("I see that the best fit is Rbf")
```
